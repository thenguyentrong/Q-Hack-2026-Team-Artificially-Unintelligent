# Database Exploration
This notebook loads the `db.sqlite` database and displays its tables as pandas DataFrames.

```python
import sqlite3
import pandas as pd

# Connect to the database
conn = sqlite3.connect('db.sqlite')

# List all tables
tables = pd.read_sql_query("SELECT name FROM sqlite_master WHERE type='table';", conn)
print("Tables in database:")
print(tables)

# Load tables into a dictionary
dfs = {}
for table in tables['name']:
    dfs[table] = pd.read_sql_query(f"SELECT * FROM {table}", conn)
    print(f"\n--- {table} --- (Shape: {dfs[table].shape})")
    display(dfs[table].head())

# Keep the connection open for further work or close it if done
# conn.close()
```

## Quick Access to Data
You can now access each table using `dfs['TableName']`. For example:

```python
if 'Company' in dfs:
    display(dfs['Company'].info())
    display(dfs['Company'].describe())
```
